```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
x_linear = np.array([[1],[2],[3],[4]])
y_linear = np.array([[3],[6],[9],[12]])
linear_model = LinearRegression()
linear_model.fit(x_linear,y_linear)
user_number = float(input('Pls Enter Ur No. => '))
result = linear_model.predict([[user_number]])
print(f'Result = {result[0][0]}')
y_predic = linear_model.predict(x_linear)
linear_mse = mean_squared_error(y_linear,y_predic)
print(f'linear_mse = {linear_mse}')
x_logistic = np.array([[5],[15],[30],[40],[45],[55]])
y_logistic = np.array([[0],[0],[0],[0],[1],[1]])
logistic_model = LogisticRegression()
logistic_model.fit(x_logistic,y_logistic)
user_number = float(input('Pls Enter Ur No. => '))
result = logistic_model.predict([[user_number]])
print(f'Result = {result[0]}')
y_logi_predict = logistic_model.predict(x_logistic)
logistic_acc = accuracy_score(y_logistic,y_logi_predict)
print(f'accuracy = {logistic_acc}')



                      
```
